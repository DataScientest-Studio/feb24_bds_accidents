### 1.1 Création du df_caracteristiques_final

```python
# Créer df_caracteristiques pour les années 2005 à 2018
# Listes des années de 2005 à 2018
annees_2005_2018 = list(range(2005, 2019))
# Liste pour stocker les DataFrames des années 2005 à 2018
df_caracteristiques_2005_2018_list = []

# Charger et fusionner les fichiers des années 2005 à 2018
for annee in annees_2005_2018:
    caracteristiques_file = f"caracteristiques_{annee}.csv"
    separator = ',' if annee != 2009 else '\t'  # Utiliser ',' sauf pour 2009 où on utilise la tabulation
    caracteristiques_df = pd.read_csv(caracteristiques_file, sep=separator, low_memory=False, encoding='latin-1')
    df_caracteristiques_2005_2018_list.append(caracteristiques_df)

# Fusionner les DataFrames des années 2005 à 2018 en un seul DataFrame
df_caracteristiques_2005_2018 = pd.concat(df_caracteristiques_2005_2018_list, ignore_index=True)

# Ajouter les fichiers des années 2019 à 2022 à df_caracteristiques en tenant compte des colonnes supplémentaires
# Listes des années de 2019 à 2022
annees_2019_2022 = list(range(2019, 2023))
# Liste pour stocker les DataFrames des années 2019 à 2022
df_caracteristiques_2019_2022_list = []

# Charger et fusionner les fichiers des années 2019 à 2022
for annee in annees_2019_2022:
    caracteristiques_file = f"caracteristiques_{annee}.csv"
    caracteristiques_df = pd.read_csv(caracteristiques_file, sep=';', low_memory=False, encoding='latin-1')
    df_caracteristiques_2019_2022_list.append(caracteristiques_df)

# Fusionner les DataFrames des années 2019 à 2022 en un seul DataFrame
df_caracteristiques_2019_2022 = pd.concat(df_caracteristiques_2019_2022_list, ignore_index=True)

# Concaténer df_caracteristiques_2005_2018 et df_caracteristiques_2019_2022 pour obtenir le DataFrame final
df_caracteristiques_final = pd.concat([df_caracteristiques_2005_2018, df_caracteristiques_2019_2022], ignore_index=True)
```

```python
print(df_caracteristiques_final.shape)
# Compter le nombre de valeurs uniques dans la colonne 'Num_Acc'
unique_num_acc_count_carac = df_caracteristiques_final['Num_Acc'].nunique()

print("Nombre de valeurs uniques pour 'Num_Acc':", unique_num_acc_count_carac)
```

```python
display(df_caracteristiques_final.head())
print(df_caracteristiques_final.shape)
```
